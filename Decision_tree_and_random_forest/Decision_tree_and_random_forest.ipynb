```python
year = pd.to_datetime(raw_df.Date).dt.year
train_df = raw_df[year<2015]
val_df = raw_df[year==2015]
test_df = raw_df[year>2015]
```

```python
print(train_df.shape)
print(test_df.shape)
print(val_df.shape)
```

```python
train_df.isna().sum().sort_values(ascending=False)
```

```python
input_cols = list(train_df.columns)[1:-1]
target_col = 'RainTomorrow'
```

```python
train_input = train_df[input_cols].copy()
train_target = train_df[target_col].copy()

test_input = test_df[input_cols].copy()
test_target = test_df[target_col].copy()

val_input = val_df[input_cols].copy()
val_target = val_df[target_col].copy()
```

```python
val_input
```

```python
numeric_cols = train_input.select_dtypes(include=np.number).columns.tolist()
```

```python
cat_cols = train_input.select_dtypes(include='object').columns.tolist()
cat_cols
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean').fit(raw_df[numeric_cols])

train_input[numeric_cols]=imputer.transform(train_input[numeric_cols])
test_input[numeric_cols]=imputer.transform(test_input[numeric_cols])
val_input[numeric_cols]=imputer.transform(val_input[numeric_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(raw_df[numeric_cols])
train_input[numeric_cols]=scaler.transform(train_input[numeric_cols])
test_input[numeric_cols]=scaler.transform(test_input[numeric_cols])
val_input[numeric_cols]=scaler.transform(val_input[numeric_cols])
```

```python
val_input.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False).fit(train_input[cat_cols])

encoded_feature_name = list(encoder.get_feature_names_out(cat_cols))
encoded_feature_name
```

```python
train_input[encoded_feature_name] = encoder.transform(train_input[cat_cols])
test_input[encoded_feature_name] = encoder.transform(test_input[cat_cols])
val_input[encoded_feature_name] = encoder.transform(val_input[cat_cols])
```

```python
test_input.shape
```

```python

X_train = train_input[numeric_cols+encoded_feature_name]
X_test = test_input[numeric_cols+encoded_feature_name]
X_val = val_input[numeric_cols+encoded_feature_name]
```

```python
def errr_f(md):
    model = DecisionTreeClassifier(max_depth=md,random_state=42).fit(X_train,train_target)
    train_rrr = 1-model.score(X_train,train_target)
    val_rrr = 1-model.score(X_val,val_target)
    return { 'max_depth':md,'train_error':train_rrr,'val_error':val_rrr}
```

```python
err_df = pd.DataFrame([errr_f(md) for md in range(1,20)])
err_df
```

```python
plt.figure()
plt.plot(err_df['max_depth'],err_df['train_error'],label='train')
plt.plot(err_df['max_depth'],err_df['val_error'],label='val')
plt.legend()
```

```python
def predict_input(model,inputs):
    input_df = pd.DataFrame([inputs])
    input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])
    input_df[encoded_feature_name] = encoder.transform(input_df[cat_cols])
    X_input = input_df[numeric_cols+encoded_feature_name]
    pred = model.predict(X_input)[0]
    prob = model.predict_proba(X_input)[0][list()]
    return pred,prob
```
